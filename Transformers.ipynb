```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")

classification = classifier("Hugging Face Transformers course is an amazing learning")
classification
```

```python
classifier(
    ["Huggingface face transformers course is amazing","feeling bad to join late to the course"]
)
```

```python
classifier = pipeline("zero-shot-classification")

classifier("hugging face course about transformers",
           candidate_labels = ["education","politics","Technology"])
```

```python
# text-Generation

generator = pipeline("text-generation")
generator("Hugging Face is s goldmine for developers. what do you think about ?")
```

```python

generator = pipeline("text-generation",model="distilgpt2")
generator("Hugging Face is s goldmine for developers. what do you think about ?",
          max_length = 30,
          num_return_sequences = 2)
```

```python
unmasker = pipeline("fill-mask")
unmasker("Hugging face transformers course about <mask>",top_k=3)
```

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("Hugging Face Course About Transformers")
```

```python
question_answer = pipeline("question-answering")
question_answer(
    question = "what is the name of the course ?",
    context = "Hugging Face Course About Transformers"
)
```

```python
summarizer = pipeline("summarization")
summarizer(
    """
   The Hugging Face course on Transformers is a comprehensive resource for understanding
    and working with transformer-based models. It covers the fundamentals of transformers,
     including self-attention and multi-head attention mechanisms, and provides hands-on
     experience with the Hugging Face transformers library. The course walks through using
     pretrained models for tasks like text classification and named entity recognition,
     fine-tuning models on custom datasets, and deploying them with frameworks like FastAPI.
      Advanced topics include efficient transformer variants (e.g., DistilBERT, ALBERT),
      distributed training, and scaling models for production. This free course is ideal
       for anyone looking to gain practical experience with modern NLP models
    """)
```

```python
translator = pipeline("translation", model="facebook/nllb-200-distilled-600M")

tamil_text = "தமிழ் ஒரு அழகான மொழி."
translation = translator(tamil_text, src_lang="tam", tgt_lang="eng")
print(translation[0]['translation_text'])
```
